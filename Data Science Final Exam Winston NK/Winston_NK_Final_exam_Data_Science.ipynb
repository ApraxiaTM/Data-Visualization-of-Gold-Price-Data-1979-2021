World Currency Ratio Dataset

```python
Country = pd.read_csv("/content/drive/MyDrive/Data Science Final Exam Winston NK/Winston NK - Country currency ratio.csv")
display(Country)
```

World Map

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.cm import ScalarMappable
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

worldmap = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))


c = Country['Longitude']
d = Country['Latitude']

mask = (Country['Currency Ratio'] < 1) | (Country['Currency Ratio'] < 10) | (Country['Currency Ratio'] < 100) | (Country['Currency Ratio'] < 1000) | (Country['Currency Ratio'] > 1000)
Country = Country[mask]

Country['color'] = np.where(Country['Currency Ratio'] < 1, 'More Valuable', 
                           np.where(Country['Currency Ratio'] < 10, 'Valuable', 
                                    np.where(Country['Currency Ratio'] < 100, 'Less Valueable', 
                                             np.where(Country['Currency Ratio'] < 1000, 'Not Valueable', 'Weak Currency'))))

color = {'More Valuable':'lawngreen',
         'Valuable':'yellow',
         'Less Valueable':'gold',
         'Not Valueable':'lightsalmon',
         'Weak Currency':'firebrick'}



sizes = [20, 40, 60, 80, 100]


fig, ax = plt.subplots(figsize=(20, 8))
worldmap.plot(color="lightgrey", ax=ax, cmap='terrain')


Map = sns.scatterplot(x=c, y=d, data=Country, hue='color', palette=color, sizes=sizes)

le.fit(Country['color'])


Country['color'] = le.transform(Country['color'])


divider = make_axes_locatable(ax)
ax_cb = inset_axes(ax, width="5%", height="50%", loc=6)


scalar_mappable = ScalarMappable(cmap='terrain')
scalar_mappable.set_array(Country['color'])

cb = fig.colorbar(scalar_mappable, cax=ax_cb)
cb.set_label('Altitude of the Land', fontsize=15)


plt.suptitle('Currency Values Around the World', fontsize=20,
             fontfamily='serif', fontweight='bold', style='italic')


plt.show()

Map.axes.get_xaxis().set_visible(False)
Map.axes.get_yaxis().set_visible(False)
```
